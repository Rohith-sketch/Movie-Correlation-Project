```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
#finding nulls in every column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df[df.isnull().any(axis=1)]
```

```python
#dropping rows with NaN
df = df.dropna(subset=['budget', 'gross','rating','company','runtime'])
df[df.isnull().any(axis=1)]
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
df.head()
```

```python
df.dtypes
```

```python
#changing datatypes
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['released'] = df['released'].astype(str)
#Extract the year using regex
df['year'] = df['released'].str.extract(r'(\d{4})')
df['year'] = df['year'].astype(int)
df
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
df.drop_duplicates()
```

```python
#corelations
#scatterplot budget vs gross 

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget VS Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.savefig('BudgetvsGross.png')
plt.show()
```

```python
df.head()
```

```python
#plot using seaborn
sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"red"} ,  line_kws={"color":"blue"})
```

```python
correlation_matrix = df.select_dtypes(include=['number']).corr()
correlation_matrix
```

```python
#High correlation between budget and gross
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlations Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.savefig('Correlations_Matrix.png')
plt.show()
```

```python
#numerize df
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlations Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```
